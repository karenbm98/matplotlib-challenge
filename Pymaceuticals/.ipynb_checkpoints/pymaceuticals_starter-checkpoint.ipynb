```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
single_df = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
single_df.head(5)
```

```python
# Checking the number of mice.
mouse_count = single_df['Mouse ID'].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = single_df.loc[single_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
single_df.loc[single_df['Mouse ID'] == 'g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
update_df = single_df[single_df['Mouse ID'].isin(duplicate_mice)==False]
update_df.head(5)
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count = update_df['Mouse ID'].nunique()
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_tumor_volume = update_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
mean_tumor_volume = mean_tumor_volume.rename('Mean Tumor Volume')
median_tumor_volume = update_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
median_tumor_volume = median_tumor_volume.rename('Median Tumor Volume')
variance_tumor_volume = update_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
variance_tumor_volume = variance_tumor_volume.rename('Tumor Volume Variance')
std_tumor_volume = update_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
std_tumor_volume = std_tumor_volume.rename('Tumor Volume Std. Dev.')
sem_tumor_volume = update_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
sem_tumor_volume = sem_tumor_volume.rename('Tumor Volume Std. Err.')

summary_df = pd.concat((mean_tumor_volume,median_tumor_volume,variance_tumor_volume,std_tumor_volume,sem_tumor_volume), axis=1)
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
df_capomulin = update_df.loc[update_df["Drug Regimen"] == "Capomulin"]
df_ramicane = update_df.loc[update_df["Drug Regimen"] == "Ramicane"]
df_infubinol = update_df.loc[update_df["Drug Regimen"] == "Infubinol"]
df_ceftamin = update_df.loc[update_df["Drug Regimen"] == "Ceftamin"]

# Start by getting the last (greatest) timepoint for each mouse
greatest_capomulin = df_capomulin.groupby('Mouse ID').max()['Timepoint']
greatest_ramicane = df_ramicane.groupby('Mouse ID').max()['Timepoint']
greatest_infubinol = df_infubinol.groupby('Mouse ID').max()['Timepoint']
greatest_ceftamin = df_ceftamin.groupby('Mouse ID').max()['Timepoint']

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
capomulin_all = pd.merge(greatest_capomulin, update_df, on=("Mouse ID","Timepoint"),how="left")
ramicane_all = pd.merge(greatest_ramicane, update_df, on=("Mouse ID","Timepoint"),how="left")
infubinol_all = pd.merge(greatest_infubinol, update_df, on=("Mouse ID","Timepoint"),how="left")
ceftamin_all = pd.merge(greatest_ceftamin, update_df, on=("Mouse ID","Timepoint"),how="left")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes
# add subset 
# Determine outliers using upper and lower bounds

treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
treatments = update_df[update_df["Drug Regimen"].isin(treatment_list)]
tumorvoldata_list = []
last_timepoint = treatments.groupby(["Drug Regimen", "Mouse ID"]).agg(tumor_size=("Tumor Volume (mm3)", lambda x: x.iloc[-1]))
last_timepoint = last_timepoint.stack(level=0).unstack(level=0)
treatments = update_df[update_df["Drug Regimen"].isin(treatment_list)]

for treatment in treatment_list:
    quartiles = last_timepoint[treatment].quantile([.25,.5,.75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
boxplot_list = []
for treatment in treatment_list:
    boxplot_list.append(list(last_timepoint[treatment].dropna()))
    
    
fig = plt.figure()
plt.xlabel("Treatment")
plt.xticks([1,2,3,4], treatment_list)
plt.ylabel("Tumor Volume")
plt.title("Tumor Volume by Treatment")
plt.boxplot(boxplot_list)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
fig1, ax1 = plt.subplots()
average_tumor_volume = df_capomulin.groupby(['Mouse ID']).mean()

marker_size=15
plt.scatter(average_tumor_volume['Weight (g)'],average_tumor_volume['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Averag Tumor Volume (mm3)')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = linregress.pearsonr(average_tumor_volume['Weight (g)'],average_tumor_volume['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

plt.scatter(average_tumor_volume["Weight (g)"],average_tumor_volume["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```
